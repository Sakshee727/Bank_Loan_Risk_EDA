### Binning

```python
# Descibing the 'AMT_INCOME_TOTAL' and 'AMT_CREDIT' columns

loan_app_data[['AMT_INCOME_TOTAL','AMT_CREDIT']].describe()
```

```python
# Lets Create a Bin for 'AMT_CREDIT' and 'AMT_INCOME_TOTAL'

amt_credit_bin = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,425000,450000,
                  475000,500000,10000000000]
amt_credit_slot = ['0-25000', '25000-50000','50000-75000','75000,100000','100000-125000', '125000-150000', '150000-175000',
                   '175000-200000','200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
                   '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above']
```

```python
# Creating the Bins 

loan_app_data['AMT_CREDIT_RANGE'] = pd.cut(loan_app_data.AMT_CREDIT , amt_credit_bin, labels = amt_credit_slot)
loan_app_data['AMT_INCOME_RANGE'] = pd.cut(loan_app_data.AMT_INCOME_TOTAL , amt_credit_bin, labels = amt_credit_slot)
```
